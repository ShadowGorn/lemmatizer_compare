Посчитаем средние значения результата лемматизации и инфлекции для каждого средства, сгруппированные по 2 изначальным формам

```python
res = y.pivot_table(index='tool', columns='type', values='result', aggfunc='mean')
rev = y.pivot_table(index='tool', columns='type', values='revert', aggfunc='mean')
```

Посчитаем общее количество слов в каждой группе, для каждого средства.

Посчитаем количество верно сверенных лемм для средства и формы

```python
res_sum = y.pivot_table(index='tool', columns='type', values='result', aggfunc='sum').T
res_sum
```

```python
diff = abs(res-rev)/res
diff.T[['PHPMorphy_inflector','SimpleNLG_inflector']]
```

Посмотрим, насколько хуже работает инфлекция и лемматизация, если правильно было только одно преобразование (из инфлекции и лемматизации)
Коэффициент считается к полному количеству преобразований

```python
err = x.copy()

#-1 = сработала только инфлекция
#0  = сработало все или ничего
#1  = сработала только лемматизация
err['only_lemm'] = err['result'] - err['revert']
err['only_inf'] = err['only_lemm']

# 1 - если сработала только инфлекция
# 0 - иначе
ind = err['only_inf'] > -1
err.loc[ind, 'only_inf'] = 0
err['only_inf'] *= -1

# 1 - если сработала только лемматизация
# 0 - иначе
ind = err['only_lemm'] < 1
err.loc[ind, 'only_lemm'] = 0


err = err.drop(['result','revert'], axis=1)

err_res = err.pivot_table(index='tool', columns='type', values=['only_lemm','only_inf'], aggfunc='mean')
err_res = err_res.stack(0).T[['PHPMorphy_inflector','SimpleNLG_inflector']]
err_res
```
