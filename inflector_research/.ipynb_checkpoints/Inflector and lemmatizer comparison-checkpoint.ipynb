Посчитаем средние показатели для каждого средства

```python
x.groupby('word_operator')['result'].mean()
```

Уберем столбцы со словами для подсчета статистики

```python
y = x.drop(['word', 'expected', 'real'], axis=1)
```

Посчитаем средние значения результата лемматизации и инфлекции для каждого средства, сгруппированные по 2 изначальным формам

```python
res = y.pivot_table(index='word_operator', columns='POS', values='result', aggfunc='mean')
```

Посчитаем общее количество слов в каждой группе, для каждого средства.

Столбцы должны быть одинаковы

```python
cnt = y.pivot_table(index='word_operator', columns='POS', values='result', aggfunc='count').T
cnt
```

Посчитаем количество верно сверенных лемм для средства и формы

```python
res_sum = y.pivot_table(index='word_operator', columns='POS', values='result', aggfunc='sum')
res_sum
```

Посмотрим результаты сравнения двух форм по лемме для каждого средства по типам форм

```python
res.T
```

```python
sns.heatmap(res, fmt="10.4f", annot=True)
```

Посчитаем, сколько преобразований в каждой категории не смогло выполнить каждое средство

```python
(cnt - res_sum.T)
```

Сделаем удобную таблицу результатов для анализа пар слов, с которыми не справляются средства

```python
exact_word = pd.crosstab([x['POS'],x['expected'],x['word']],[x['result'],x['word_operator']])[1]
exact_word
```

Для взятия нужных строк/столбцов в такой структуре необходимо изпользовать мультииндекс

```python
idx = pd.IndexSlice
exact_word.loc[idx[['VBD','VBN'],:,:],idx['NLTK_lemmatizer','PHPMorphy_inflector']]
```

Распрямим таблицу для выборки по столбцам

```python
plane_exact_word = exact_word.reset_index()
plane_exact_word
```

Для взятия суффикса у нужного столбца 

```python
plane_exact_word['word'].str[-3:]
```

Сохраним результат в csv файл

```python
plane_exact_word['suff'] = plane_exact_word['word'].str[-3:]
```

```python
plane_exact_word.groupby(['POS', 'suff']).sum()
```

Отберем другую выборку, сделав отсечение по суффиксу

```python
need = (plane_exact_word['type'] == 'VBP/VBZ') & (plane_exact_word['first'].str[-2:] == 'se') & (plane_exact_word['PHPMorphy_inflector'] == 1)
plane_exact_word[need]
```

И сохраним выборку в csv

```python
plane_exact_word[need][['first','second']].to_csv('chosen_data')
```
